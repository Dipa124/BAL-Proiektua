```python
import random as r
import pandas as pd
```

```python
data = {
    "LEVEL":[level],
    "MAX GRADE":[max_grade], 
    "AVG GRADE":[avg_grade],  
    "HEIGHT":[height], 
    "WEIGHT":[weight],
    "GENDER":[gender],
    "AGE":[age], 
    "TIME":[time], 
    "STYLE":[style], 
    "HOLD":[hold]}
```

```python
styles = ["Dynamic", "Static", "Acrobatic", "Mix"]
hold_types = ["Jug", "Crimp", "Sloper", "Pinch", "Pocket", "Crack", "Volume", "Mono", "Edge"]
genders = ["Male", "Female"]
levels = ["Beg", "Int", "Adv", "Pro", "Eli"]
grades = {
    "Beg": ["4", "5", "5+", "6a"],
    "Int": ["6a+", "6b", "6b+", "6c", "6c+", "7a"],
    "Adv": ["7a+", "7b", "7b+", "7c"],
    "Pro": ["7c+", "8a", "8a+", "8b"],
    "Eli": ["8b+", "8c", "8c+", "9a", "9a+", "9b", "9b+", "9c"]
}
```

```python
valid_num = False
while valid_num != True:
    data_amount = input("Ingrese el numero de datos deseado: ")
    try:
        data_amount = int(data_amount)
        valid_num = True
    except:
        pass

df = pd.DataFrame({"LEVEL":[],
                   "MAX GRADE":[], 
                   "AVG GRADE":[],  
                   "HEIGHT":[], 
                   "WEIGHT":[],
                   "GENDER":[],
                   "AGE":[], 
                   "TIME":[], 
                   "STYLE":[], 
                   "HOLD":[]})

while data_amount > 0:
    level = r.choice(levels)
    max_grade = r.choice(grades[level])
    
    grades_normal = r.randint(0, 4)
    if level == "Beg":
        if (grades[level].index(max_grade)-grades_normal) >= 0:
            avg_grade = grades[level][grades[level].index(max_grade)-grades_normal]
        if (grades[level].index(max_grade)-1) >= 0:
            avg_grade = grades[level][grades[level].index(max_grade)-1]
        else:
            avg_grade = max_grade
    else:
        if (grades[level].index(max_grade)-grades_normal) >= 0:
            avg_grade = grades[level][grades[level].index(max_grade)-grades_normal]
        else:
            avg_grade = grades[levels[levels.index(level)-1]][-r.randint(1, 2)]
   
    height = r.randint(160, 195)
    weight = r.randint(55, 90)
    gender = r.choice(genders)
    age = r.randint(10, 60)        
    time = r.randint(0, age-5)
    style = r.choice(styles)
    hold = r.choice(hold_types)
    
    nr = pd.DataFrame({"LEVEL":[level],
                       "MAX GRADE":[max_grade], 
                       "AVG GRADE":[avg_grade],  
                       "HEIGHT":[height], 
                       "WEIGHT":[weight],
                       "GENDER":[gender],
                       "AGE":[age], 
                       "TIME":[time], 
                       "STYLE":[style], 
                       "HOLD":[hold]})
    
    df = pd.concat([df, nr],
                 axis=0,
                 ignore_index=True)
    
    data_amount -= 1
```
