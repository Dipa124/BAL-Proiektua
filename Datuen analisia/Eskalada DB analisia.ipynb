```python
# DEFINITU TEXTUA ZENBAKIETAN
aldatuta_eskaladaDB = eskaladaDB.copy()

# DEFINITU SEXUAK ZENBAKIETAN
sexuak_zenbakietan = {
    "Emakumea": 0,
    "Gizona": 1
}

aldatuta_eskaladaDB["SEXUA"] = aldatuta_eskaladaDB["SEXUA"].map(sexuak_zenbakietan).astype(int)

# IZENAK ETA ABIZENAK KENDU
aldatuta_eskaladaDB = aldatuta_eskaladaDB.drop(["IZENA", "ABIZENA"], axis=1)

# DEFINITU ADIN TARTEA ZENBAKIETAN
adin_tartea_zenbakietan = {
    "Nerabea": 0,
    "Gaztea": 1,
    "Heldua": 2
}

aldatuta_eskaladaDB["ADIN TARTEA"] = aldatuta_eskaladaDB["ADIN TARTEA"].map(adin_tartea_zenbakietan).astype(int)

# DEFINITU MAILA ZENBAKIETAN
maila_zenbakietan = {
    "Hasiberria": 0,
    "Erdikoa": 1,
    "Aurreratua": 2,
    "Profesionala": 3,
    "Elitea": 4
}

aldatuta_eskaladaDB["MAILA"] = aldatuta_eskaladaDB["MAILA"].map(maila_zenbakietan).astype(int)

# DEFINITU ESTILOA ZENBAKIETAN
estiloa_zenbakietan = {
    "Dinamikoa": 0,
    "Estatikoa": 1,
    "Nahasia": 2
}

aldatuta_eskaladaDB["ESTILOA"] = aldatuta_eskaladaDB["ESTILOA"].map(estiloa_zenbakietan).astype(int)

# DEFINITU PRESAK ZENBAKIETAN
presak = ["Jug", "Crimp", "Sloper", "Pinch", "Pocket", "Crack", "Volume", "Mono", "Edge"]
presen_luzeera = len(presak)
presen_luzeeraren_zerrenda = []
while presen_luzeera > 0:
    presen_luzeeraren_zerrenda.append(len(presak)-presen_luzeera)
    presen_luzeera -= 1
presak_zenbakietan = dict(zip(presak, presen_luzeeraren_zerrenda))

aldatuta_eskaladaDB["PRESA"] = aldatuta_eskaladaDB["PRESA"].map(presak_zenbakietan).astype(int)

# ERAKUTSI DATUAK ALDATUTA
aldatuta_eskaladaDB
```
