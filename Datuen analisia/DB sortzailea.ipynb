**DATUAK SORTZEN DITUZTEN FUTZIOAK**

```python
def Sexua():
    probabilitatea = r.randint(1, 10)
    if probabilitatea <= 6:
        return "Gizona"
    elif probabilitatea >= 7:
        return "Emakumea"
sexua = Sexua()
```

```python
izenak = r"C:\Users\dipas\Desktop\Python\Personal projects\Data analytics\BAL proiektua\Izenak.ods"
izenak_giz = pd.read_excel(izenak)["GIZONAK"].tolist()
izenak_ema = pd.read_excel(izenak)["EMAKUMEAK"].tolist()
def Izena(sexua):
    if sexua == "Gizona":
        return r.choice(izenak_giz)
    elif sexua == "Emakumea":
        return r.choice(izenak_ema)

izena = Izena(sexua)
```

```python
abizenak = pd.read_excel(r"C:\Users\dipas\Desktop\Python\Personal projects\Data analytics\BAL proiektua\Abizenak.ods")["ABIZENAK"].tolist()
def Abizena():
    return r.choice(abizenak)

abizena = Abizena()
```

```python
def Adina():
    probabilitatea = r.randint(1, 10)
    if probabilitatea <= 3:
        return [r.randint(14, 20), "Nerabea"]
    elif probabilitatea <= 8:
        return [r.randint(21, 35), "Gaztea"]
    elif probabilitatea >= 9:
        return [r.randint(36, 55), "Heldua"]

adina, adin_tartea = Adina()
```

```python
def Eskarmentua(adina):
    probabilitatea = r.randint(1, 10)
    if probabilitatea <= 3:
        if adina <= 18:
            return r.randint(6, adina-8)
        elif adina <= 20:
            return r.randint(6, adina-10)
        elif adina > 20:
            return r.randint(6, 10)
    elif probabilitatea <= 9:
        return r.randint(0, 5)
    elif probabilitatea == 10:
        if adina <= 18:
            return r.randint(11, adina-3)
        elif adina <= 20:
            return r.randint(6, adina-8)
        elif adina > 20:
            return r.randint(11, adina-4)

eskarmentua = Eskarmentua(adina)
```

```python
def Altuera(sexua, adin_tartea):
    probabilitatea = r.randint(1, 10)
    if sexua == "Gizona":
        if adin_tartea == "Nerabea":
            if probabilitatea <= 2:
                return r.randint(160, 165)
            elif probabilitatea <= 8:
                return r.randint(165, 175)
            elif probabilitatea <= 10:
                return r.randint(175, 185)
        elif adin_tartea == "Gaztea":
            if probabilitatea <= 2:
                return r.randint(160, 170)
            elif probabilitatea <= 8:
                return r.randint(170, 185)
            elif probabilitatea <= 10:
                return r.randint(186, 195)
        elif adin_tartea == "Heldua":
            if probabilitatea <= 2:
                return r.randint(160, 170)
            elif probabilitatea <= 8:
                return r.randint(170, 180)
            elif probabilitatea <= 10:
                return r.randint(180, 190)
    elif sexua == "Emakumea":
        if adin_tartea == "Nerabea":
            if probabilitatea <= 2:
                return r.randint(145, 155)
            elif probabilitatea <= 8:
                return r.randint(155, 165)
            elif probabilitatea <= 10:
                return r.randint(165, 170)
        elif adin_tartea == "Gaztea":
            if probabilitatea <= 2:
                return r.randint(150, 160)
            elif probabilitatea <= 8:
                return r.randint(160, 170)
            elif probabilitatea <= 10:
                return r.randint(170, 180)
        elif adin_tartea == "Heldua":
            if probabilitatea <= 2:
                return r.randint(150, 160)
            elif probabilitatea <= 8:
                return r.randint(160, 165)
            elif probabilitatea <= 10:
                return r.randint(165, 175)
            
adina = Altuera(sexua, adin_tartea)
```

```python
def Masa(sexua, adin_tartea):
    probabilitatea = r.randint(1, 10)
    if sexua == "Gizona":
        if adin_tartea == "Nerabea":
            if probabilitatea <= 3:
                return r.uniform(50, 60)
            elif probabilitatea <= 8:
                return r.uniform(60, 70)
            else:
                return r.uniform(70, 80)
        elif adin_tartea == "Gaztea":
            if probabilitatea <= 3:
                return r.uniform(60, 70)
            elif probabilitatea <= 8:
                return r.uniform(70, 80)
            else:
                return r.uniform(80, 90)
        elif adin_tartea == "Heldua":
            if probabilitatea <= 3:
                return r.uniform(70, 80)
            elif probabilitatea <= 8:
                return r.uniform(75, 85)
            else:
                return r.uniform(80, 90)
    elif sexua == "Emakumea":
        if adin_tartea == "Nerabea":
            if probabilitatea <= 3:
                return r.uniform(45, 55)
            elif probabilitatea <= 8:
                return r.uniform(55, 65)
            else:
                return r.uniform(65, 75)
        elif adin_tartea == "Gaztea":
            if probabilitatea <= 3:
                return r.uniform(55, 65)
            elif probabilitatea <= 8:
                return r.uniform(65, 75)
            else:
                return r.uniform(75, 85)
        elif adin_tartea == "Heldua":
            if probabilitatea <= 3:
                return r.uniform(65, 75)
            elif probabilitatea <= 8:
                return r.uniform(70, 80)
            else:
                return r.uniform(75, 85)
            
masa = Masa(sexua, adin_tartea)
```

```python
mailak = ["Hasiberria", "Erdikoa", "Aurreratua", "Profesionala", "Elitea"]
def Maila(sexua, eskarmentua):
    probabilitatea = r.randint(1, 100)
    if sexua == "Gizona":
        if eskarmentua >= 11:
            if probabilitatea <= 8:
                return "Elitea"
            elif 8 < probabilitatea <= 38:
                return "Profesionala"
            elif 38 < probabilitatea <= 93:
                return "Aurreratua"
            else:
                return "Erdikoa"
        elif eskarmentua >= 6:
            if probabilitatea <= 12:
                return "Elitea"
            elif 12 < probabilitatea <= 34:
                return "Profesionala"
            elif 34 < probabilitatea <= 84:
                return "Aurreratua"
            else:
                return "Erdikoa"
        elif eskarmentua >= 2:
            if probabilitatea <= 8:
                return "Profesionala"
            elif 8 < probabilitatea <= 40:
                return "Aurreratua"
            elif 48 < probabilitatea <= 98:
                return "Erdikoa"
            else:
                return "Hasiberria"
        elif eskarmentua <= 1:
            if probabilitatea <= 8:
                return "Aurreratua"
            elif 8 < probabilitatea <= 75:
                return "Hasiberria"
            else:
                return "Erdikoa"
    
    elif sexua == "Emakumea":
        if eskarmentua >= 11:
            if probabilitatea <= 6:
                return "Elitea"
            elif 6 < probabilitatea <= 31:
                return "Profesionala"
            elif 31 < probabilitatea <= 86:
                return "Aurreratua"
            else:
                return "Erdikoa"
        elif eskarmentua >= 6:
            if probabilitatea <= 10:
                return "Elitea"
            elif 10 < probabilitatea <= 30:
                return "Profesionala"
            elif 30 < probabilitatea <= 80:
                return "Aurreratua"
            else:
                return "Erdikoa"
        elif eskarmentua >= 2:
            if probabilitatea <= 6:
                return "Profesionala"
            elif 6 < probabilitatea <= 42:
                return "Aurreratua"
            elif 42 < probabilitatea <= 96:
                return "Erdikoa"
            else:
                return "Hasiberria"
        elif eskarmentua <= 1:
            if probabilitatea <= 7:
                return "Aurreratua"
            elif 7 < probabilitatea <= 82:
                return "Hasiberria"
            else:
                return "Erdikoa"

            
maila = Maila(sexua, eskarmentua)
```

```python
def Estiloa():
    return r.choice(["Dinamikoa", "Estatikoa", "Nahasia"])

estiloa = Estiloa()
```

```python
def Presa():
    probabilitatea = r.randint(1, 10)
    if probabilitatea <= 3:
        return r.choice(["Crimp", "Sloper", "Pinch", "Pocket", "Crack", "Volume"])
    elif probabilitatea <= 8:
        return r.choice(["Jug", "Crimp", "Sloper", "Pinch"])
    elif probabilitatea >= 9:
        return r.choice(["Mono", "Edge"])

presa = Presa()
```

**DATU BASE SORTZAILEA**

```python
zenbaki_baliogarria = False
while zenbaki_baliogarria != True:
    datu_kopurua = input("Ezarri datu kopurua: ")
    try:
        datu_kopurua = int(datu_kopurua)
        zenbaki_baliogarria = True
    except:
        pass

df = pd.DataFrame({"SEXUA":[],
                   "IZENA":[], 
                   "ABIZENA":[],  
                   "ADINA":[], 
                   "ADIN TARTEA":[],
                   "ESKARMENTUA":[],
                   "ALTUERA":[], 
                   "MASA":[], 
                   "MAILA":[], 
                   "ESTILOA":[],
                   "PRESA":[]})

while datu_kopurua > 0:
    sexua = Sexua()
    izena = Izena(sexua)
    abizena = Abizena()
    adina, adin_tartea = Adina()
    eskarmentua = Eskarmentua(adina)
    altuera = Altuera(sexua, adin_tartea)
    masa = round(Masa(sexua, adin_tartea), 2)
    maila = Maila(sexua, eskarmentua)
    estiloa = Estiloa()
    presa = Presa()
    
    nr = pd.DataFrame({"SEXUA":[sexua],
                       "IZENA":[izena], 
                       "ABIZENA":[abizena],  
                       "ADINA":[adina], 
                       "ADIN TARTEA":[adin_tartea],
                       "ESKARMENTUA":[eskarmentua],
                       "ALTUERA":[altuera], 
                       "MASA":[masa], 
                       "MAILA":[maila], 
                       "ESTILOA":[estiloa],
                       "PRESA":[presa]})
    
    df = pd.concat([df, nr],
                 axis=0,
                 ignore_index=True)
    
    datu_kopurua -= 1
```
